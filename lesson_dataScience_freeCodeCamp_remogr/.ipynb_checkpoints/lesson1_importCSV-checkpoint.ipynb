# importing CSV and analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# loading data

```python
sales = pd.read_csv('data/sales_data.csv', parse_dates=['Date'])
```

```python
sales['Profit'].mean() #can use min, std, max etc in the describe field
```

```python
sales['Unit_Price'].plot(kind='box', vert=False, figsize=(14,6))
```

```python
sales['Unit_Price'].plot(kind='density', figsize=(14,6))
```

```python
ax = sales['Unit_Price'].plot(kind='density', figsize=(14,6))
ax.axvline(sales['Unit_Price'].mean(), color = 'red')
ax.axvline(sales['Unit_Price'].median(), color = 'green')
```

```python
pax = sales['Unit_Price'].plot(kind='hist', figsize=(14,6))
pax.set_ylabel('Number of Sales')
pax.set_xlabel('Dollars')
```

```python
sales['Age_Group'].value_counts()
```

```python
sales['Age_Group'].value_counts().plot(kind = 'pie', figsize=(6,6))
```

```python
sales['Age_Group'].value_counts().plot(kind = 'bar', figsize=(12,5))
```

```python
sales.plot(kind = 'scatter', x = 'Customer_Age', y = 'Profit', figsize = (12,5))
```

```python
fig = sales[['Age_Group', 'Profit']].boxplot(by = 'Age_Group', figsize = (12,5))
fig.set_ylabel('Profit')
```

```python
cols = ['Customer_Age', 'Profit', 'Unit_Price', 'Cost']
sales[cols].plot(kind = 'box', subplots = True, layout=(1,4), figsize = (15,5))
```
